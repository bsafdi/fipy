```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import sys,os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib import rcParams
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rcParams['text.usetex'] = True
rcParams['text.latex.unicode'] = True

from fipy import gaussian
```

# Make Curve to filter

```python
def f(x):
    return np.exp(-np.abs(x)/10.0)

x_range = np.linspace(-20,20.0,50)
f_range = f(x_range)

plt.plot(x_range,f_range)
```

## The filter

We'll take two different filters.  In the first case, we will use a constant $\sigma$ array, in the second we will use an array where $\sigma$ depends on $x$

```python
sigma_range = np.abs(x_range)/2.0
sigma_range_const = np.ones(50)
```

```python
plt.plot(x_range,sigma_range_const,'k-',label="consant sigma")
plt.plot(x_range,sigma_range,'b--',label="variable sigma")
plt.legend()
plt.xlabel("x")
plt.ylabel(r"$\sigma$")
```

# Filter curve with variable $\sigma$'s

```python
f_range_filtered = gaussian.gaussian_filter1d(x_range,f_range,sigma_range)


plt.plot(x_range,f_range,'k-',label="unfiltered")
plt.plot(x_range,f_range_filtered,'b--',label="filtered")
plt.legend()
```

```python
print "check they both sum to the same value: ", np.sum(f_range)/np.sum(f_range_filtered)
```
